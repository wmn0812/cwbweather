## 載入需要用到的套件

```python
import json
import requests
import pandas as pd
from pandas.core.frame import DataFrame
from bs4 import BeautifulSoup
```

## crawling

```python
url = 'http://www.cwb.gov.tw/V7/forecast/week/week.htm'
resp = requests.get(url) 
resp.encoding = 'utf-8'
soup = BeautifulSoup(resp.text, 'lxml')
tbody = soup.find_all("tr") #解析內文

all_text_df =[] #作為存放所有內容的list

name = ["地區"] #補一個地區的值
name_list = tbody[0].find_all("th")[1:] #撈取最上面的標題，但第一個不是我們要的故去掉

for a in range(len(name_list)): #將網頁語法去除掉
    name.append(name_list[a].text)  
```

```python
for i in range(len(tbody)): #撈取內文部分
    try: #撈到一半時中間會因為換地區而參雜星期，故這邊先做一次例外處理，把不是地區的值給排除掉
        clear_text = [] #裝一個地區的欄位
        try: #基隆市有兩個，但網頁上面是將兩個表格併在一起，所以要寫兩次的話這邊要設第二個例外處理
            clear_text.append(tbody[i].find("th",{"rowspan":"2"}).text)
        except:
            clear_text.append(tbody[i-1].find("th",{"rowspan":"2"}).text)
        #print(clear_text)
        text = tbody[i].find_all("td") #地區以外的欄位
        for n in range(len(text)): #清理網頁語法並裝到陣列裡面
            try:
                title = text[n].find("img")["title"]
            except:
                title = ""
            data = title + text[n].text.strip()
            clear_text.append(data)
        all_text_df.append(clear_text) #將每個地區都裝到大表格裡面去
    except:
        pass
```

```python
all_df = pd.DataFrame(all_text_df, columns = name) #將二維陣列變成dataframe的形式
all_df
# all_df.to_json('./cwbweather.json',force_ascii=False,orient = 'columns')
```
